```python
import matplotlib.pyplot as plt

# Creating a horizontal bar chart
kc_df.plot(kind='barh', x='bedrooms', y='floors', legend=False) 
#col1 provides the categories on the y-axis and col2 provides the values

# Adding titles and labels
plt.xlabel("Values")
plt.ylabel("Number of bedrooms")
plt.title("Horizontal Bar Chart of col1 vs. col2")

plt.show()
```

```python
bedroom_counts = kc_df['bedrooms'].value_counts().sort_index()

# Creating a horizontal bar chart
bedroom_counts.plot(kind='barh', color='skyblue', legend=False)

# Adding titles and labels
plt.xlabel("Number of properties")
plt.ylabel("Number of bedrooms")
plt.title("Horizontal Bar Chart of Number of Properties by Bedroom Count")

plt.show()
```
